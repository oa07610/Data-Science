```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
df = pd.read_csv('AutoMPG.csv')
```

```python
df
```

### Unique value columns such as unique id, unique names should not be used in Machine Learning models

```python
df = df.drop(['car name'],axis=1)
```

```python
df
```

### Convention or Rule of Thumb is to call/denoted all the inputs as X and all the output/prediction/response as Y

```python
X = df.drop(['mpg'],axis=1)
Y = df['mpg']
```

```python
X
```

```python
Y
```

### Simple Linear Regression between mpg and weight

Using just two variables. One input column and one output/prediction/response column. The reason behind using one input and one output is just to learn linear regression and see the model (straight line) on the 2-D plot

Coding Trick: Need to add constant everytime with your input. The reason for this is that it takes care of the bias (the intercept term b
 is often called the bias parameter) in the data (a constant difference which is there for all observations).

```python
X = sm.add_constant(X['weight'])
```

```python
X
```

```python
model = sm.OLS(Y,X).fit()
```

```python
model.summary()
```

```python
model.pvalues
```

### Interpretation of the above result

R-squared: 0.669 (any value above 0.3 usually shows good relationship between variables) <br>
Y (prediction) = MPG <br> X (input) = weight <br> For each additional pound/kg of weight on the car (or for every unit increase in weight),
the MPG decreases by 0.0076 miles

```python
plt.plot(df['weight'].values, df['mpg'].values,'ro')
plt.plot(df['weight'].values ,model.fittedvalues.values,  'b')
plt.legend(['Data', 'Fitted model'])
plt.xlabel('weight')
plt.ylabel('mpg')
```

## Regression can't work with categorical inputs

### One Hot Encoding - For converting categorical columns into numerical columns

```python
df = pd.read_csv('AutoMPG.csv')
```

```python
df
```

### Just creating a categorical column to show how one hot encoding works

```python
df['Type'] = 'Old'
```

```python
df.head()
```

```python
df.loc[df['model year'] > 70, 'Type'] = 'New'
```

```python
df = df.drop('car name',axis=1)
```

```python
df
```

```python
X = df.drop(['mpg'],axis=1)
Y = df['mpg']
```

```python
X.head()
```

```python
Y
```

```python
X = sm.add_constant(X)
```

```python
X.head()
```

```python
modelcat = sm.OLS(Y,X).fit()
```

### We get the error above because Regression cannot process categorical columns
### Lets try one hot encoding

```python
one_hot = pd.get_dummies(df, columns = ['Type'], dtype=int)
```

```python
one_hot
```

```python
X = one_hot.drop(['mpg'],axis=1)
Y = one_hot['mpg']
```

```python
X = sm.add_constant(X)
```

```python
X.head()
```

```python
modelcat = sm.OLS(Y,X).fit()
```

```python
modelcat.summary()
```

### Why we fit models?
### Because we want to predict values (output) for new/unknown things (inputs)

### Lets save the model first in a file

```python
# Save the model to disk
filename = 'regmodelcat.pkl'
pickle.dump(modelcat, open(filename, 'wb'))
```

### Lets load the model first in a file

```python
# Load the model from disk
loaded_model = pickle.load(open('regmodelcat.pkl', 'rb'))
```

As of now, you can use either model_cat or loaded_model since they both are present in this notebook. For prediction on other notebooks, you can skip the fit() process and just load the saved model to do predictions. You dont even have to read the csv data file.

### Lets create one new input (car) and predict its output (mpg)

```python
df_new = pd.DataFrame({'const': [1.0],
                       'cylinders':[8],
                       'displacement':[307],
                       'horsepower': [130],
                       'weight': [3504],
                       'acceleration': [12],
                       'model year':[70],
                       'origin': [1],
                       'Type_New': [0],
                       'Type_Old':[1]})
```

```python
df_new
```

### We have already trained a model above. No need to train again. Just predict

```python
#predicting mpg for new data using modelcat
modelcat.predict(df_new)
```

```python
#predicting mpg for new data using loaded_model
loaded_model.predict(df_new)
```

### Interpretation

#### For the new car, the predicted MPG is 16.77 using our regression model
