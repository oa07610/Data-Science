```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('loan_small.csv')
df=df.drop_duplicates()
```

```python
df['term'] = df['term'].str.replace('months','')
df['term'] = df['term'].astype(int)
```

```python
df['loan_status']
```

```python
df['loan_status'] = df['loan_status'].replace(['Fully Paid'], 'Excellent')
df['loan_status'] = df['loan_status'].replace(['Charged Off'], 'Very Bad')
df['loan_status'] = df['loan_status'].replace(['Current'], 'Average')
df['loan_status'] = df['loan_status'].replace(['Default'], 'Very Bad')
df['loan_status'] = df['loan_status'].replace(['Late (31-120 days)'], 'Risky')
df['loan_status']
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['emp_title', 'emp_length', 'last_pymnt_d'])
df = df.drop(['mths_since_last_delinq', 'mths_since_last_record'],axis=1)
df.head()
```

```python
print(df['emp_length'].unique())
```

```python
df['emp_length'] = df['emp_length'].replace(['10+ years'], '14')
df['emp_length'] = df['emp_length'].str.replace('years', '')
df['emp_length'] = df['emp_length'].str.replace('year', '')
df['emp_length'] = df['emp_length'].str.replace('< 1', '0')
df['emp_length'] = df['emp_length'].astype(int)
```

```python
df['verification_status'] = df['verification_status'].replace(['Source Verified', 'Verified'], 'v')
df['verification_status'] = df['verification_status'].str.replace('Not Verified', 'nv')
```

```python
df['funded_amnt_inv'] = df['funded_amnt_inv'].astype(int)
df['annual_inc'] = df['annual_inc'].astype(int)
```

```python
df.columns = [i.capitalize() for i in df.columns]
df.to_csv('cleaned_loan_small.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
%matplotlib inline
```

```python
df = pd.read_csv('cleaned_loan_small.csv')
df.head()
```

```python
#df['Annual_inc'].describe()
```

```python
df.groupby('Loan_status')['Annual_inc'].describe()
```

```python
#By using the above analysis, we can see that the status of most loans are Excellent while some some are average. There is also a significant portion that is very bad too which indicates that many people were/have not been able to pay off their loans. Most of the people that tend to have a salary of more than 60000, it can be said that their loan stasuses are usually good, indicating that income plays a major role in the loan payment.
```

```python
df.groupby(['Grade', 'Loan_status'])['Loan_amnt'].describe()
```

```python
#one interesting thing visible from this data was the fact that most of the loan amount was always near the 20k mark. The most loans were given to the category A and a surprising discovery would be the fact that the highest mean from all grades (in excellent category of loan status) is of G.
```

```python
# using Verfication Status for visualization
sns.countplot(x='Verification_status', data=df)
plt.title('Count of Sources Verified')
```

```python
#the graph above shows that most of the loans that were given were from a verified source, meaning that the bank make sures that its customers are people that have a history of being good creditors.
```

```python
sns.histplot(df['Annual_inc'], kde=True, bins=15)
plt.title('Histogram of Annual Income Amount')
```

```python
#the graph above shows that most of the people that recieved the loan had an income of around 50000 or greater than it, showing that the bank focuses on checking current financial conditions before giving loans.
```

```python
sns.scatterplot(x='Loan_amnt', y='Int_rate', data=df)
plt.title('Scatter plot of Amount of Loan vs Interest Rate')
```

```python
#using the scatter plot for the analysis, we can see that there is a positive corelation between the amount of loan given and the interest charged by the bank on it. This makes sense as a bank would charge a greater interest on a greater amount of loan as there is a greater risk involved when giving big loans, hence a greater return policy that is adopted by the bank in this scenario.
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=df,x='Grade', y='Loan_amnt')
```

```python
#the above box plot shows that the F has recieved the biggest amount of loan, and also that it had aa great variety of loans offered, from small to big. Funny thing to not here is that Grade A has less loan amount than others.
```

```python
sns.catplot(x="Grade", hue="Loan_status", col="Verification_status",
                data=df, kind="count",
                height=6, aspect=1);
plt.title('Multivariate Analysis of Loan Status, Grade, and Verification Status')
```

```python
#this analysis of the grade, verification status, and loan status shows us that the people that have a verified source tend to have a better payment record which is shown by the greater heights of the excellent category in the Verified part of the graph.
```
