```python
q1 = qns.iloc[:,[0,1]]
q1.columns = q1.iloc[0].astype(str)
q1 = q1[1:]
q1 = q1.replace(1, pd.Series(q1.columns, q1.columns)).fillna('')
q1 = pd.DataFrame(q1.iloc[:,0] + q1.iloc[:,1])
q1.columns = ['Q1']

q1.head()
```

```python
q2 = qns.iloc[:,[2,3,4,5]]
q2.columns = q2.iloc[0].astype(str)
q2 = q2.iloc[1:]

for i in q2.columns:
    q2[i] = q2[i].replace('1', i).fillna('')

q2 = pd.DataFrame(q2.iloc[:,0] + q2.iloc[:,1] + q2.iloc[:,2] + q2.iloc[:,3])
q2.columns = ['Q2']

q2.head()
```

```python
q3_7 = qns.iloc[:,[6,7,8,9,10]]
q3_7 = q3_7[1:]

q3_7.head()
```

```python
# getting q8 columns
q8 = qns.iloc[:, [11,12,13,14]]
q8.columns = q8.iloc[0].astype(str)
q8 = q8[1:]

for i in q8.columns:
    q8[i] = q8[i].replace(['0', '1', '2'], i).fillna('')

q8 = pd.DataFrame(q8.iloc[:,0] + q8.iloc[:,1] + q8.iloc[:,2] + q8.iloc[:,3])
q8.columns = ['Q8']

q8.head()
```

```python
# getting q9
q9 = qns.iloc[:, [15,16,17]]
q9.columns = q9.iloc[0].astype(str)
q9 = q9[1:]

for i in q9.columns:
    if i == 'Yes, Full Time':
        q9[i] = q9[i].replace('1', 'Working Full Time').fillna('')
    else:
        q9[i] = q9[i].replace(['0', '1', '2'], 'Working Part Time' if i == 'Working, but not Full Time' else 'Not Working').fillna('')

q9 = pd.DataFrame(q9.iloc[:,0] + q9.iloc[:,1] + q9.iloc[:,2])
q9.columns = ['Q9']

q9.head()
```

```python
#Q10
q10 = qns.iloc[:, [18,19,20,21,22]]
q10.columns = q10.iloc[0].astype(str)
q10 = q10[1:]

for i in q10.columns:
    q10[i] = q10[i].replace('1', i).fillna('')
q10 = pd.DataFrame(q10.iloc[:,0] + q10.iloc[:,1] + q10.iloc[:,2] + q10.iloc[:,3] + q10.iloc[:,4])
q10.columns = ['Q10']

q10.head()
```

```python
# Q11
q11 = qns.iloc[:, [23,24,25,26,27]]
q11.columns = q11.iloc[0].astype(str)
q11 = q11[1:]

for i in q11.columns:
    q11[i] = q11[i].replace('1', i).fillna('')

q11 = pd.DataFrame(q11.iloc[:,0] + q11.iloc[:,1] + q11.iloc[:,2] + q11.iloc[:,3] + q11.iloc[:,4])
q11.columns = ['Q11']

q11.head()
```

```python
# Q12
q12 = qns.iloc[:, [28,29,30,31,32]]
q12.columns = q12.iloc[0].astype(str)
q12 = q12[1:]

for i in q12.columns:
    q12[i] = q12[i].replace('1', i).fillna('')

q12 = pd.DataFrame(q12.iloc[:,0] + q12.iloc[:,1] + q12.iloc[:,2] + q12.iloc[:,3] + q12.iloc[:,4])
q12.columns = ['Q12']

q12.head()
```

```python
# Q13 
q13 = qns.iloc[:, [33,34]]
q13.columns = q13.iloc[0].astype(str)
q13 = q13[1:]

for i in q13.columns:
    q13[i] = q13[i].replace('1', i).fillna('')

q13 = pd.DataFrame(q13.iloc[:,0] + q13.iloc[:,1])
q13.columns = ['Q13']

q13.head()
```

```python
# Q14 
q14 = qns.iloc[:, [35,36,37]]
q14.columns = q14.iloc[0].astype(str)
q14 = q14[1:]

for i in q14.columns:
    q14[i] = q14[i].replace('1', i).fillna('')

q14 = pd.DataFrame(q14.iloc[:,0] + q14.iloc[:,1] + q14.iloc[:,2])
q14.columns = ['Q14']

q14.head()
```

```python
q15 = qns.iloc[:, [38]]
q15.columns = ['Q15']
q15 = q15[1:].fillna('Empty')
q15.head()
```

```python
total = pd.concat([data,q1,q2,q3_7,q8,q9,q10,q11,q12,q13,q14,q15],axis=1)

total.to_csv('formatted_580SurveyCleanupFinal.csv', index = False)

total.head()
```

## Part 2

```python
f_dataframe = pd.read_csv('formatted_580SurveyCleanupFinal.csv')
f_dataframe.isnull().sum()
```

```python
pd.set_option('mode.chained_assignment',  None)
```

```python
#Dropping columns that have empty values
f_dataframe = f_dataframe.drop(['Access Code', 'Email Address'], axis=1)
```

```python
#Fixing Q3 and using mode or the most frequently occurring value to fill in missing values
f_dataframe['Q3'] = f_dataframe['Q3'].str.replace(' ', '')
f_dataframe['Q3'] = f_dataframe['Q3'].replace('1993', '28')
f_dataframe['Q3'].fillna(f_dataframe['Q3'].mode()[0],inplace=True)

#f_dataframe.loc[:,'Q3']
```

```python
#Using mode or the most frequently occurring value to fill in missing values
f_dataframe['Region'] = f_dataframe['Region'].fillna(f_dataframe['Region'].mode()[0])
f_dataframe['City / Town / District'] = f_dataframe['City / Town / District'].fillna(f_dataframe['City / Town / District'].mode()[0])
```

```python
#Replacing incorrect values of Q4: Height
f_dataframe['Q4'] = f_dataframe['Q4'].replace({'7O':'70', '6l':'61', '5.2':'52'})

#Missing Value of Q4 HEight
f_dataframe['Q4'] = pd.to_numeric(f_dataframe['Q4'], errors='coerce')
f_dataframe['Q4'] = f_dataframe['Q4'].fillna(f_dataframe['Q4'].mean())

#f_dataframe.loc[:,'Q4']
```

```python
f_dataframe['Q6'] = f_dataframe['Q6'].str.title().str.strip().replace({
    'Bachelor of engineering ': 'Engineering',
    'Engineering ': 'Engineering',
    'Cse': 'Computer Science and Engineering',
    'Computer science engineering ': 'Computer Science and Engineering',
    'Computer engineering': 'Computer Science and Engineering',
    'B.Tech In Cs': 'Computer Science',
    'Criminology law and soci': 'Criminology law and social sciences',
    'Ece': 'Electrical and Computer Engineering',
    'Economic': 'Economics',
    'Bit': 'Business in Technology',
    'Yes': None,
})

f_dataframe['Q6'].fillna(f_dataframe['Q6'].mode()[0],inplace=True)
#f_dataframe.head()
```

```python
f_dataframe['Q8'] = f_dataframe['Q8'].replace('Microsoft/Windows', 'Windows')
f_dataframe['Q8'] = f_dataframe['Q8'].replace('Apple/MacBook', 'Macbook')
f_dataframe['Q8'] = f_dataframe['Q8'].replace('Microsoft/WindowsApple/MacBook', 'Both')
f_dataframe['Q8'] = f_dataframe['Q8'].replace('OtherAcer/Windows', 'Windows')

f_dataframe['Q15'] = f_dataframe['Q15'].replace('Empty', 'Not Responded')

f_dataframe.head()
```

```python
f_dataframe['Q5'] = f_dataframe['Q5'].str.strip()
f_dataframe['Q5'] = f_dataframe['Q5'].str.title()
f_dataframe['Q5'] = f_dataframe['Q5'].replace(['Usa', 'United states', 'United states ', 'Us'], 'United States')
f_dataframe['Q5'] = f_dataframe['Q5'].replace(['United States ; Pakistan'], 'United States and Pakistan')

#Not filling Missing Values as it will make the data significantly inaccurate. A good approximation can be using mode but based on the town/region.

f_dataframe.loc[:,'Q5']
```

```python
#Filling the rest of the columns with string

string_columns = ['Q7','Q8','Q9','Q10','Q11','Q12']

for i in string_columns:
    f_dataframe[i].fillna('Not responded', inplace=True)

f_dataframe
```

```python
f_dataframe.to_csv('cleaned_580SurveyCleanupFinal.csv', index = False)

f_dataframe
```
